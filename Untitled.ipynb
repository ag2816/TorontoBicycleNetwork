## Convert the To and From stations to edges in NetworkX

```python
ride_edges = nx.from_pandas_edgelist(rides,'from_station_name','to_station_name',edge_attr=None)
```

```python
layout = nx.kamada_kawai_layout(ride_edges)

plt.figure(figsize=(10,10))
plt.axis('off')
nx.draw_networkx_nodes(ride_edges, layout, node_color='darkgray', node_size=120, alpha=.75)
nx.draw_networkx_edges(ride_edges, layout, edge_color='darkgray',width=1, alpha=.75)
```

```python
nx.draw(ride_edges,with_labels=True)
```

# centrality
* The degree of a node is simply defined as the number of connecting edges that it has
* in our case, the higher the centrality, the more in- and out-bound trips to that location
* "Popularity"

```python
degree_centrality = nx.degree_centrality(ride_edges)
degree_centrality
```

```python
import operator

def centrality_sort(centrality_dict):
    return sorted(centrality_dict.items(),key=operator.itemgetter(1))
```

```python
#centrality : which nodes have the highest /lowest degree centrality

degree_cent = nx.degree_centrality(ride_edges)
degree_sorted = centrality_sort(degree_cent)

print('----------------------Degree Centrality--------------------------')
print('Highest degree:',degree_sorted[-5:])
print('\n')

print('Lowest degree:',degree_sorted[:5])
```

```python

#betweenness centrality : which nodes have the highest /lowest betweenness centrality

between_cent = nx.betweenness_centrality(ride_edges)
between_sorted = centrality_sort(between_cent)

print('----------------------Betwenness Centrality--------------------------')
print('Highest degree:',between_sorted[-5:])
print('\n')

print('Lowest degree:',between_sorted[:5])
```

```python
#closenness centrality : which nodes have the highest and lowest closeness centrality

closeness_cent = nx.closeness_centrality(ride_edges)
closeness_sorted = centrality_sort(closeness_cent)

print('----------------------Closenness Centrality--------------------------')
print('Highest degree:',closeness_sorted[-5:])
print('\n')

print('Lowest degree:',closeness_sorted[:5])
```

```python
highest_degree = [node[0] for node in degree_sorted[-20:]]
```

```python
#create a subgraph
sub = ride_edges.subgraph(highest_degree)
nx.draw(sub,with_labels=True)
```

```python
lowest_degree = [node[0] for node in degree_sorted[:20]]
```

```python

sub_low = ride_edges.subgraph(lowest_degree)
nx.draw(sub_low,with_labels=True,)
```

```python
plt.figure(figsize=(10,10))
plt.axis('off')

sub_low = ride_edges.subgraph(lowest_degree)
nx.draw(sub_low,with_labels=True,node_size=120, alpha=.75)
```

## Eigenvector
* The basic idea behind this metric revolves around a nodes neighbors and how connected they are. To score higher, a node needs to be well connected (high degree centrality) but it also needs to be connected to others that are well connected.
* An interpretation of this metric, Influence.

```python

#nx.adjacency_matrix(ride_edges).todense()
eigenvector_centrality = nx.eigenvector_centrality(ride_edges)
eigenvector_centrality
```

```python
eigen_sorted = centrality_sort(eigenvector_centrality)
highest_degree = [node[0] for node in eigen_sorted[-20:]]
#create a subgraph
sub = ride_edges.subgraph(highest_degree)
nx.draw(sub,with_labels=True)
```
